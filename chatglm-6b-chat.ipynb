```python
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("../autodl-tmp/chatglm-6b", trust_remote_code=True)
model = AutoModel.from_pretrained("../autodl-tmp/chatglm-6b", trust_remote_code=True,local_files_only=True).half().cuda()
response, history = model.chat(tokenizer, "你好", history=[])
print(response)
response, history = model.chat(tokenizer, "晚上睡不着应该怎么办", history=history)
print(response)
```
